```python
import pandas as pd
import os
data = pd.read_csv("Myntra_Women_Clothing_Dataset/Myntra_Women_Clothing_Dataset _lessSize.csv")
patterns = ['Checked','Woven','Embroidered','Printed','Striped']
fabrics = ['Cotton','Polyester','Nylon','Synthetic','Modal']

```

```python
data.Pattern.dropna
data.Fabric.dropna
```

```python
data = data.sample(frac = 1)
data[:5]
```

```python
import urllib
import requests
try:
    count = 2246
    for i,dat in data[2246:].iterrows():
        imges = dat['Image_Urls'].strip('][').split(', ')
        pat = dat['Pattern']
        fab = dat['Fabric']
        col = dat['Primary_Colour']
        col = col.replace(" ","_")
        if pat in patterns or fab in fabrics:
            for img in imges:
                response = requests.get(img[1:-1],headers={"user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) snap Chromium/74.0.3729.169 Chrome/74.0.3729.169 Safari/537.36"})
                file = open(f"images-1/{pat}-{fab}-{col}-{count}.jpg", "wb")
                file.write(response.content)
                file.close()
                count+=1
except:
    pass
```
